## 二分法
- 左闭右闭
- 左闭右开

**remark:** 注意left和right初始化，while循环语句的条件

````python
from IPython.testing.decorators import skipif


class Solution1 :#左闭右闭
    def Binarysearch1(self,nums,val):
        left,right = 0 , len(nums)-1

        while left <= right:
            mid = (left + right) // 2

            if nums[mid]<val:
                left = mid + 1
            elif nums[mid]>val:
                right = mid - 1
            else:
                return mid
        return -1 #没有找到

class Solution2:#左闭右开
    def BinarySearch2(self,nums,val):
         left,right = 0,len(nums)

         while left < right:
             mid = (left + right) // 2
             if nums[mid] > val:
                 right = mid
             elif nums[mid] < val:
                 left = mid + 1
             else:
                 return mid
         return -1
````

## 移除元素
题目大意：
给你一个数组 nums 和一个值 val，你需要 原地 移除所有数值等于 val 的元素，并返回移除后数组的新长度。

方法：快慢指针

````python
class Solution3:
    def remove_elements(self, nums, val):
        #快慢指针
        slow,fast,size = 0,0,len(nums)
        while fast < size:
             if nums[fast] != val:
                 nums[slow] = nums[fast]
                 slow += 1
             fast += 1
        return slow

    def remove_elements2(self,nums,val):
        i,size = 0,len(nums)
        while i < size:
            if nums[i] == val:
                for j in range(i+1,size):
                    nums[j] = nums[j-1]
                size -= 1
                i -= 1 #注意这里的
            i += 1
        return i
````

## 有序数组的平方

题目大意：给你一个按 非递减顺序 排序的整数数组 nums，返回 每个数字的平方 组成的新数组，要求也按 非递减顺序 排序。

当然除了solution4双指针之外，还可以进行进一步优化
- 优化所有元素同号的情况
- 头尾平方的大小直接相加与0比较即可
- 插入列表尾部的索引可以通过for循环来实现

````python
class Solution4:#双指针法+反转列表法
    def sortedSquares(self,nums):
        left,right = 0,len(nums)-1
        result = []
        while left <= right:
            if abs(nums[left]) < abs(nums[right]):
                result.append(nums[right] ** 2)
                right -= 1
            else:
                result.appned(nums[left] ** 2)
                left += 1
        return result[::-1]

    def sortedSquares2(self,nums):#双指针
        l,r,i = 0,len(nums)-1,len(nums)-1
        result = [0]*len(nums)
        while l <= r:
            if abs(nums[l]) < abs(nums[r]):
                result[i] = nums[r] ** 2
                r -= 1
            else :
                result[i] = nums[l] ** 2
                l += 1
            i -= 1
        return result

    def sortedSquares3(self,nums):#暴力排序+列表生成法
        return sorted(x*x for x in nums)
    #使用 sorted([x*x for x in nums])也是可以的，但是后者需要先生成一个列表，再进行排序，而前者不需要，因此节省了内存。当数据较多时候，推荐使用前者
````

## 长度最小的字数组
题目大意：给定一个含有 n 个正整数的数组和一个正整数 s ，找出该数组中满足其和 ≥ s 的长度最小的 连续 子数组，并返回其长度。如果不存在符合条件的子数组，返回 0。


````python
class Solution5:#暴力算法
    def minsubarray(self,nums,val):
        min_len = 0
        for i in range(len(nums)):
            sum = 0
            for j in range(i,len(nums)):
                sum += nums[j]
                if sum >= val:
                    min_len = min(min_len,j-i+1)
                    break
        return min_len

    def minsubarray2(self,nums,val): #滑动窗口
        i , sum , min_len = 0,0,0

        for j in range(len(nums)):
            sum += nums[j]
            while sum >= val:
                min_len = min(min_len,j-i+1)
                sum -= nums[i]
                i +=1
        return min_len
````

## 区间和
类似ACM输入输出模式

题目大意：给定一个整数数组 Array，请计算该数组在每个指定区间内元素的总和

输入描述：第一行输入为整数数组 Array 的长度 n，接下来 n 行，每行一个整数，表示数组的元素。随后的输入为需要计算总和的区间，直至文件结束。

输出表述：输出每个指定区间内元素的总和。

示例如下：
```
5
1
2
3
4
5
0 1
1 3

```


````python
import sys
input = sys.stdin.read
class Solution6:
    def main(self):
        data = input().split()
        index = 0
        n = int(data[index])
        index += 1
        sum = 0 #累积和
        lst = []
        for i in range(n):
            sum += int(data[index])
            lst.append(sum)
            index += 1
        while index < len(data):
            a  = int(data[index])
            b  = int(data[index+1])

            if a == 0:
                print(lst[b])
            else:
                print(lst[b]-lst[a-1])

            index += 2
````

## 开发商购买土地

**题目描述：**

  在一个城市区域内，被划分成了n * m个连续的区块，每个区块都拥有不同的权值，代表着其土地价值。目前，有两家开发公司，A 公司和 B 公司，希望购买这个城市区域的土地。

  现在，需要将这个城市区域的所有区块分配给 A 公司和 B 公司。

  然而，由于城市规划的限制，只允许将区域按横向或纵向划分成两个子区域，而且每个子区域都必须包含一个或多个区块。

  为了确保公平竞争，你需要找到一种分配方式，使得 A 公司和 B 公司各自的子区域内的土地总价值之差最小。

  注意：区块不可再分。

**输入描述：**

  第一行输入两个正整数，代表 n 和 m。

  接下来的 n 行，每行输出 m 个正整数。

**输出描述：**

  请输出一个整数，代表两个子区域内土地总价值之间的最小差距。

````python
def main():
    import sys
    input = sys.stdin.read
    data = input().split()

    idx = 0
    n = int(data[idx])
    idx += 1
    m = int(data[idx])
    idx += 1
    sum = 0
    vec = []
    for i in range(n):
        row = []
        for j in range(m):
            num = int(data[idx])
            idx += 1
            row.append(num)
            sum += num
        vec.append(row)

    # 统计横向
    horizontal = [0] * n
    for i in range(n):
        for j in range(m):
            horizontal[i] += vec[i][j]

    # 统计纵向
    vertical = [0] * m
    for j in range(m):
        for i in range(n):
            vertical[j] += vec[i][j]

    result = float('inf')
    horizontalCut = 0
    for i in range(n):
        horizontalCut += horizontal[i]
        result = min(result, abs(sum - 2 * horizontalCut))

    verticalCut = 0
    for j in range(m):
        verticalCut += vertical[j]
        result = min(result, abs(sum - 2 * verticalCut))

    print(result)

if __name__ == "__main__":
    main()
````
